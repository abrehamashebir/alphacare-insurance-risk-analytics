```python
gender_specified = df[df['Gender'] != 'Not specified']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=gender_specified,x='Gender',y='TotalPremium', palette='Set2')
plt.title('Total Premium Per Gender"')
plt.xlabel('Gender')
plt.ylabel('Total Premium')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../images/total_premium')
plt.show()
```
